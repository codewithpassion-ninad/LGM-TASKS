Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import imageio
import scipy.ndimage
```

**Loading and displaying the image**

```python
image = cv2.imread("D:\OneDrive\Pictures\Screenshots\Screenshot 2023-08-24 235307.png")
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

**Converting RGB image to Gray image**

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
```

**Converting a gray image to inverted image**

```python
inverted_image =255-gray_image
plt.imshow(inverted_image)
```

**Further, converting image into blurred image using gaussian blur function**

```python
blurred_image=cv2.GaussianBlur(inverted_image,(21,21),0)
plt.imshow(blurred_image)
```

**Getting the inverted blurred image from the blurred image**

```python
inverted_blurred=255-blurred_image
pencil_sketch=cv2.divide(gray_image,inverted_blurred,scale=256.0)
plt.imshow(pencil_sketch)
```

**Compiling all the above steps in one function**

```python
def convert_to_sketch(img):
    grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    invert_img = cv2.bitwise_not(grey_img)
    blur_img = cv2.GaussianBlur(invert_img, (111, 111), 0)
    invblur_img = cv2.bitwise_not(blur_img)
    sketch_img = cv2.divide(grey_img, invblur_img, scale=256.0)
    cv2.imwrite('sketch.jpg', sketch_img)
    return sketch_img
```

**Retreiving the sketch of the image feeded to the function**

```python
sketch_img = convert_to_sketch(image)
```

**Displaying the original image and the sketch image parallely**

```python
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
plt.imshow(RGB_image)
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(sketch_img, cmap='gray')
plt.axis('off')
plt.title('Sketch')

plt.show()
```
